**2   Write a code to plot multiple axes in one figure**

```python
import numpy as np

# we are defining a function to produce some numbers, given a value of t
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)


# t1 is range of numbers from 0 to 5 , incrementing by 0.1
# code t1 here
t1=np.arange(0.0,5.0,0.1)

# t2 is range of numbers from 0 to 5 , incrementing by 0.02 
# code t2 here
t2=np.arange(0.0,5.0,0.02)

```

```python
plt.figure()
plt.subplot(2, 1, 1) # no of subplot,col,row

plt.plot(t1, f(t1), 'bo')  # Note : x values are t1 and y values are f(t1)
plt.ylabel("Y label-1");

plt.subplot(2, 1, 2)  # so 2,1,2 means 2 rows , 1 column and we want to plot in the 2nd cell.
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')  # Note : x values are t2 and y values are np.cos(2*np.pi*t2)

plt.ylabel("Y label-2"); # Which subplot is modifying this function?
# Ans : the current one. i.e subplot (2, 1, 2)
```

## Object Oriented Interface

**3. Demostrtate object oriented interface,  to create the objects and then call methods for every object to make changes specifically to that object. The `plt.subplots` function creates a figure object and an Subplot (Axes) object at the same time, then we can create a plot by calling methods on those objects.**

```python
x = np.linspace(-4,4)  # default num = 50 values
y1, y2 = x**2, x**3
```

```python
# Object Oriented Interface
fig, ax = plt.subplots()
ax.plot(x, y1, 'red')
ax.plot(x, y2, 'blue')
ax.set_title('Quadratic and Cubic Powers')
ax.set_xlabel('x values')
ax.set_ylabel('y values'); # put a ; to remove the unnecessary Text(...) part
```

```python
# pyplot
plt.plot(x, y1, 'red', x, y2, 'blue')
plt.title('Quadratic and Cubic Powers')
plt.xlabel('x values')
plt.ylabel('y values');
```

<font color='green'><b>The usefulness of the OOI is clearer when working with multiple subplots in one figure.</b></font> We've mentioned before that a figure can have more than one `Axes` on it.  If you want your axes to be on a regular grid system, then it's easiest to use `plt.subplots(...)` to create a figure and add the axes to it automatically.

```python
# create a figure object with with 4 subplots as 2 by 2
fig,axes=plt.subplots(nrows=2,ncols=2)
```

`plt.subplots(...)` created a new figure and added 4 subplots to it.  The `axes` object that was returned is a 2D numpy object array.  Each item in the array is one of the subplots.  They're laid out as you see them on the figure. Therefore, when we want to work with one of these axes, we can index the `axes` array and use that item's methods. For example:

**3 Creating a figure showing the shape of the power functions using 1) one axes 2) using subplots**

```python
x = np.linspace(start=-5, stop=5, num=150)
```

```python
# All functions in one axes
fig, ax = plt.subplots(figsize = (7,4))
ax.plot(x, x, label='Linear')
ax.plot(x, x**2,  label='Quadratic')
ax.plot(x, x**3, label='Cubic')
ax.plot(x, x**4, label='4th Power')
ax.legend();

# Notice : linear and Quadratic are hardly seen in their real shapes.
# becoz they are being overpowered by higher order equations
```

```python
# 4 Axes/Subplots: One function in one axes
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (7,7))
axes[0,0].set_title('Linear')
axes[0,0].plot(x,x)
## more code here
axes[0,1].set_title('Quadratic')
axes[0,1].plot(x,x**2)
axes[1,0].set_title('Cubic')
axes[1,0].plot(x,x**3)
axes[1,1].set_title('Quadratic')
axes[1,1].plot(x,x**4)


```

**6. Using Bar graph show comparision between gold and silver medal in Rio olympics**

**7.use a graphical representation to sense density of distribution of data**

```python
iqs = np.random.normal(loc=100, scale=10, size=300)
# np.random.normal produces a normal distribution around 100 of 300 points
# scale = 10 means standard deviation 

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

ax[0].hist(iqs, bins=28, edgecolor='k')
ax[0].set_title('Frequency histogram')
    
ax[1].hist(iqs, bins=18, color = 'red', edgecolor='k', density=True)
ax[1].set_title('Density (normed) histogram');
```
